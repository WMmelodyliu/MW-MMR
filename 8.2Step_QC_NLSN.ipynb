```python
import pandas as pd
import glob
import numpy as np
import os
```

```python
#old
old_numerator = pd.read_csv("MMR_Numerator_FY26Feb_Total_nlsn2(414).csv") # full deliverable

#new
new_numerator = pd.read_csv("FY26Mar_mmr_numerator_bq_nlsn2.csv")
```

```python
old_numerator.head()
```

```python
new_numerator.tail()
```

```python
#old version
prev_mon = 'Feb' ## older numerator(v1)
#new version
curr_mon = 'Mar' ## new numerator (v2)
```

```python
#check 1
date_check_1 = '202501'
#check 2
date_check_2 = '202502'
```

```python
old_numerator['version'] = prev_mon
new_numerator['version'] = curr_mon
```

```python
#concatenate calendar_year and calendar_month
#old
for index, row in old_numerator.iterrows():
    if len(str(old_numerator.loc[index,'cal_month_nbr']))==1:
        old_numerator.loc[index,'concat_date'] = str( old_numerator.loc[index,'cal_year_nbr'])+"0"+str( old_numerator.loc[index,'cal_month_nbr'])
    else:
        old_numerator.loc[index,'concat_date'] = str( old_numerator.loc[index,'cal_year_nbr'])+str( old_numerator.loc[index,'cal_month_nbr'])
#new
for index, row in new_numerator.iterrows():
    if len(str(new_numerator.loc[index,'cal_month_nbr']))==1:
        new_numerator.loc[index,'concat_date'] = str(new_numerator.loc[index,'cal_year_nbr'])+"0"+str(new_numerator.loc[index,'cal_month_nbr'])
    else:
        new_numerator.loc[index,'concat_date'] = str(new_numerator.loc[index,'cal_year_nbr'])+str(new_numerator.loc[index,'cal_month_nbr'])
```

```python
#append old and new
#df = old_numerator.append(new_numerator)
df = pd.concat([old_numerator, new_numerator])
```

```python
df.head()
```

```python
#check month
print(prev_mon," date",df[df['version']==prev_mon]['concat_date'].unique())
print(curr_mon, " date",df[df['version']==curr_mon]['concat_date'].unique())
```

### QC:  compare sales difference % and $ over period

```python
#need attention if |$diff| > 5M or |diff%| > 1%
def highlight_red_or_green(val):
    color = 'red' if abs(val) >5000000 else 'green'
    return 'color: %s' % color
```

#### Offline over periods

```python
offline_QC_period = pd.pivot_table(df[df['Channel']=='Offline'], values='WM_Sales', index=['cal_year_nbr', 'cal_month_nbr'],
                    columns=['version'], aggfunc=np.sum, fill_value=0)
offline_QC_period.head()
#column name need to be changed
offline_QC_period["diff"] = offline_QC_period[curr_mon] - offline_QC_period[prev_mon]
offline_QC_period["diff%"] = (offline_QC_period[curr_mon] - offline_QC_period[prev_mon])/offline_QC_period[prev_mon]

offline_QC_period.style.format({
    curr_mon: '{:,.2f}'.format,
    prev_mon: '{:,.2f}'.format,
    'diff': '{:,.2f}'.format,
    'diff%': '{:,.2%}'.format,
})
```

```python
offline_QC_period[['diff','diff%']].style.applymap(highlight_red_or_green)
```
